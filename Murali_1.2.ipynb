```r
library("ggplot2")
library("ggmap")
library("maptools")
library("maps")
states<-map_data("state")
#states
heatmap = read.csv("StateDatabyWeekforMap_2017-18week40-4.csv")
heatmap$region = (tolower(heatmap$STATENAME))
#heatmap$ACTIVITY.LEVEL = as.integer(gsub("Level ", "", heatmap$ACTIVITY.LEVEL))
#heatmap
fr=c()

for(i in heatmap$ACTIVITY.LEVEL){
  if (i=="Level 0"){
    fr=c(fr,0)
  }
  if (i=="Level 1"){
    fr=c(fr,1)
  }
  if (i=="Level 2"){
    fr=c(fr,2)
  }
  if (i=="Level 3"){
    fr=c(fr,3)
  }
  if (i=="Level 4"){
    fr=c(fr,4)
  }
  if (i=="Level 5"){
    fr=c(fr,5)
  }
  if (i=="Level 6"){
    fr=c(fr,6)
  }
  if (i=="Level 7"){
    fr=c(fr,7)
  }
  if (i=="Level 8"){
    fr=c(fr,8)
  }
  if (i=="Level 9"){
    fr=c(fr,9)
  }
  if (i=="Level 10"){
    fr=c(fr,10)
  }
}
heatmap$ACTIVITY.LEVEL=fr
merged<- merge(states,heatmap, by="region", all.x=T)
#merged
ggplot(merged, aes(x=long, y=lat, group=group))+
  geom_polygon(aes(fill=ACTIVITY.LEVEL))+
  geom_path()+ 
  scale_fill_gradientn(colours=c("green4", "green3","green", "yellowgreen", "yellow2",  "yellow", "darkgoldenrod1", "orange", "orangered", "red"),na.value="white")+
  coord_map()
```
